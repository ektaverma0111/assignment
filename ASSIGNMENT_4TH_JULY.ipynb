# 1 = What is a random variable in probability theory ?
Ans = In probability theory, a random variable is a fundamental concept used to quantify outcomes of a random phenomenon.
A random variable is a function that assigns a numerical value to each outcome in a sample space of a random experiment.

Types of Random Variables:

Discrete Random Variable:

Takes on countable values (like integers).

Example: Number of heads in 3 coin tosses (values: 0, 1, 2, 3).

Continuous Random Variable:

Takes on uncountable values, typically real numbers in an interval.

Example: The time a customer waits in line (values: any real number ≥ 0).

Key Properties:

Domain: All possible outcomes of an experiment.

Range: All numerical values the variable can take.

Probability Distribution:

For discrete variables: Defined using a Probability Mass Function (PMF).

For continuous variables: Defined using a Probability Density Function (PDF).

A random variable bridges the gap between abstract outcomes of a random process and numerical analysis using mathematics and statistics. It helps in calculating probabilities, expected values, variances, and making statistical inferences.

Ans = In probability theory, random variables are classified based on the type of values they can take. The two main types are:

# 3 =  What is the difference between discrete and continuous distributions ?

Ans = The difference between discrete and continuous distributions lies in the type of random variable they describe, the way probabilities are represented, and how values are measured or counted.

1. Discrete Distributions

Definition:

A discrete distribution deals with a discrete random variable, which takes on specific, countable values.

Key Characteristics:

Values are separated and distinct (e.g., 0, 1, 2, ...).

Represented using a Probability Mass Function (PMF).

The probability of each individual value can be greater than zero.

The total probability of all values is exactly 1.

Examples:

Binomial distribution: Number of successes in a fixed number of trials.

Poisson distribution: Number of events in a fixed interval.

Geometric distribution: Number of trials until first success.

2. Continuous Distributions

Definition:

A continuous distribution deals with a continuous random variable, which can take any value within a range or interval.

Key Characteristics:

Values are infinite and uncountable (e.g., all real numbers between 0 and 1).

Represented using a Probability Density Function (PDF).

The probability of any single exact value is zero.

Probabilities are found by calculating the area under the curve over an interval.

The total area under the curve is 1.

Examples:

Normal distribution: Bell-shaped curve, used for natural phenomena.

Exponential distribution: Time between events in a Poisson process.

Uniform distribution: All values in an interval are equally likely.



# 4 = What are probability distribution functions (PDF) ?

Ans = A Probability Distribution Function is a mathematical function that defines how the probabilities are distributed over the values of a random variable. It represents the likelihood of different outcomes in a probability experiment.

The type of probability distribution function depends on the nature of the random variable, which may be either discrete or continuous.

1. Discrete Probability Distribution Function

For a discrete random variable, the probability distribution function is called the Probability Mass Function (PMF).

It assigns probabilities to each possible outcome that the random variable can take.

The PMF must satisfy two conditions:

The probability of each outcome is non-negative.

The sum of the probabilities of all possible outcomes is equal to 1.

This function gives the exact probability of a discrete outcome.

2. Continuous Probability Distribution Function

For a continuous random variable, the probability distribution function is called the Probability Density Function (PDF).

It defines a curve such that the area under the curve within a certain interval represents the probability that the random variable lies within that interval.

For a continuous variable, the probability of any exact value is zero. Probabilities are only meaningful over intervals.

The total area under the entire curve of the PDF equals 1.

# 21 =  What are the assumptions for applying the Central Limit Theorem ?

Ans = The Central Limit Theorem (CLT) allows us to approximate the sampling distribution of the sample mean by a normal distribution, even if the population distribution is not normal. However, certain theoretical assumptions must be met for the CLT to apply reliably.

1. Independent Observations

The sampled data must be independent.

Each observation should not influence or depend on the others.

This is typically ensured by random sampling from the population.

Why important?
Independence prevents systematic patterns that could bias the sampling distribution.

2. Identically Distributed Variables

The observations should be drawn from the same population (i.e., identically distributed).

This means that each value comes from the same probability distribution with the same mean and variance.

Why important?
To ensure the sample statistics are representative of a single distribution, not a mixture.

3. Finite Mean and Variance

The population from which samples are drawn must have a finite mean (μ) and finite variance (σ²).

The CLT does not apply if the variance is infinite or undefined (as in some heavy-tailed distributions).

Why important?
Without a defined center and spread, the normal approximation breaks down.

4. Sufficiently Large Sample Size

As sample size

```python
#3= Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution ?
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Define a function to compute the PDF of Bernoulli distribution
def bernoulli_pdf(p):
    """
    This function calculates and displays the PDF (Probability Distribution Function)
    of a Bernoulli distribution with probability of success `p`.
    """
    # Valid outcomes: 0 (failure), 1 (success)
    x = [0, 1]

    # Compute the PDF values using scipy
    pdf = bernoulli.pmf(x, p)

    # Display the PDF values
    for xi, pi in zip(x, pdf):
        print(f"P(X = {xi}) = {pi:.2f}")

    # Plot the PDF
    plt.figure(figsize=(5, 4))
    plt.bar(x, pdf, tick_label=["0 (Fail)", "1 (Success)"], color=["red", "green"])
    plt.title(f"PDF of Bernoulli Distribution (p = {p})")
    plt.xlabel("Outcome")
    plt.ylabel("Probability")
    plt.ylim(0, 1)
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.show()

# Example: Call the function with p = 0.7
bernoulli_pdf(0.7)
```

```python
# 9 = Write a Python function to calculate Z-scores from a dataset and plot them ?

# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the function to calculate Z-scores
def calculate_and_plot_z_scores(data):
    """
    This function calculates the Z-scores for a dataset
    and plots them using a bar chart.
    """
    mean = np.mean(data)
    std_dev = np.std(data)

    z_scores = (data - mean) / std_dev

    # Print Z-scores
    print("Z-scores:")
    for i, z in enumerate(z_scores):
        print(f"Data[{i}] = {data[i]}, Z = {z:.2f}")

    # Plotting Z-scores
    plt.figure(figsize=(10, 4))
    plt.bar(range(len(data)), z_scores, color='skyblue', edgecolor='black')
    plt.axhline(0, color='red', linestyle='--', linewidth=1)
    plt.title("Z-scores of Dataset")
    plt.xlabel("Index")
    plt.ylabel("Z-score")
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()

    return z_scores

# Example dataset
dataset = np.array([10, 12, 9, 15, 14, 13, 11, 10])

# Call the function with the example data
z_scores = calculate_and_plot_z_scores(dataset)
```

```python
# 10 = Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.?

# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Function to demonstrate CLT with different sample sizes
def clt_demo(sample_size=5, num_samples=1000, population_size=100000, dist='exponential'):
    np.random.seed(42)

    # Create non-normal population
    if dist == 'exponential':
        population = np.random.exponential(scale=1.0, size=population_size)
    elif dist == 'uniform':
        population = np.random.uniform(low=0, high=10, size=population_size)
    else:
        raise ValueError("Supported distributions: 'exponential', 'uniform'")

    # Generate sample means
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=False)
        sample_means.append(np.mean(sample))

    # Plot original population and sample means
    plt.figure(figsize=(12, 5))

    # Plot 1: Original Distribution
    plt.subplot(1, 2, 1)
    plt.hist(population, bins=50, color='salmon', edgecolor='black', alpha=0.7, density=True)
    plt.title(f"Original Distribution ({dist.capitalize()})")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.grid(True, linestyle="--", alpha=0.6)

    # Plot 2: Sampling Distribution of the Mean
    plt.subplot(1, 2, 2)
    plt.hist(sample_means, bins=30, color='lightblue', edgecolor='black', alpha=0.7, density=True)
    plt.title(f"Sampling Distribution of the Mean\nSample Size = {sample_size}")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.grid(True, linestyle="--", alpha=0.6)

    plt.tight_layout()
    plt.show()

# Example: Try with sample size 5, 30, and 100
clt_demo(sample_size=5, dist='exponential')
clt_demo(sample_size=30, dist='exponential')
clt_demo(sample_size=100, dist='exponential')
```

```python
# 16 =Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1) ?

# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to plot standard normal distribution with shaded regions
def plot_standard_normal_with_shading():
    """
    Plots the standard normal distribution (mean = 0, std = 1)
    and shades area under curve between Z = -1 and Z = 1.
    """
    # Generate x values
    x = np.linspace(-4, 4, 1000)

    # PDF of standard normal distribution
    pdf = norm.pdf(x, loc=0, scale=1)

    # Plot the standard normal curve
    plt.figure(figsize=(10, 5))
    plt.plot(x, pdf, 'b-', linewidth=2, label='Standard Normal PDF')

    # Shade area between Z = -1 and Z = 1
    x_shade = np.linspace(-1, 1, 500)
    plt.fill_between(x_shade, norm.pdf(x_shade), color='skyblue', alpha=0.6, label='Area: Z ∈ [-1, 1]')

    # Optional: Add lines for Z = ±1, ±2, ±3
    for z, color in zip([1, 2, 3], ['green', 'orange', 'red']):
        plt.axvline(z, color=color, linestyle='--', label=f'Z = {z}')
        plt.axvline(-z, color=color, linestyle='--', label=f'Z = -{z}')

    # Plot settings
    plt.title("Standard Normal Distribution with Shaded Area (Z ∈ [-1, 1])")
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend(loc='upper left')
    plt.show()

    # Call the function
plot_standard_normal_with_shading()
```

```python
# 18 =  Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution?
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to calculate Z-score and visualize
def calculate_and_plot_z(data_point, data_mean, data_std):
    """
    Calculates Z-score for a data point and plots it on the standard normal curve.
    """
    # Calculate Z-score
    z = (data_point - data_mean) / data_std
    print(f"Z-score for data point {data_point} (mean={data_mean}, std={data_std}): {z:.2f}")

    # Plot the standard normal distribution
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Standard Normal PDF', color='blue')

    # Shade the area under the curve up to the Z-score
    x_fill = np.linspace(-4, z, 500)
    y_fill = norm.pdf(x_fill)
    plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5, label=f'Area left of Z = {z:.2f}')

    # Vertical line for Z-score
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')

    # Plot settings
    plt.title("Z-score on Standard Normal Distribution")
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

    return z

# Example usage
data_point = 72
data_mean = 65
data_std = 10

# Call the function
z = calculate_and_plot_z(data_point, data_mean, data_std)
```

```python
# 19 = Implement hypothesis testing using Z-statistics for a sample dataset ?

# Import required libraries
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Step 1: Define your sample data
sample = np.array([52, 55, 48, 50, 47, 53, 49, 51, 46, 54])
sample_mean = np.mean(sample)
sample_size = len(sample)

# Step 2: Define population parameters
population_mean = 50   # μ0 (Null hypothesis mean)
population_std = 5     # Known population standard deviation (σ)

# Step 3: State Hypotheses
# H0: μ = 50 (null hypothesis)
# H1: μ ≠ 50 (two-tailed test)

# Step 4: Calculate Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-statistic = {z_stat:.2f}")

# Step 5: Compute p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
print(f"P-value = {p_value:.4f}")

# Step 6: Decision at 5% significance level (α = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H0). There is significant evidence.")
else:
    print("Fail to reject the null hypothesis (H0). No significant evidence.")

# Step 7: Visualize the Z-test
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)
z_critical = norm.ppf(1 - alpha/2)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Standard Normal Distribution", color='blue')

# Shade rejection regions
plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label="Rejection Region (α = 0.05)")

# Mark Z-statistic
plt.axvline(z_stat, color='black', linestyle='--', label=f"Z = {z_stat:.2f}")
plt.title("Z-Test for One Sample Mean")
plt.xlabel("Z-value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
#20= Create a confidence interval for a dataset using Python and interpret the result ?
# Import required libraries
import numpy as np
from scipy.stats import norm

# Step 1: Define the dataset
data = np.array([52, 55, 48, 50, 47, 53, 49, 51, 46, 54])
sample_mean = np.mean(data)
sample_size = len(data)
population_std = 5  # Assume known population standard deviation

# Step 2: Set confidence level
confidence_level = 0.95
alpha = 1 - confidence_level

# Step 3: Compute the Z-critical value
z_critical = norm.ppf(1 - alpha/2)

# Step 4: Compute the margin of error
margin_of_error = z_critical * (population_std / np.sqrt(sample_size))

# Step 5: Calculate confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Display results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Confidence Level = {confidence_level * 100:.0f}%")
print(f"Z-critical value = {z_critical:.2f}")
print(f"Margin of Error = {margin_of_error:.2f}")
print(f"Confidence Interval = ({lower_bound:.2f}, {upper_bound:.2f})")

# Interpretation
print("\nInterpretation:")
print(f"We are {int(confidence_level*100)}% confident that the true population mean lies between {lower_bound:.2f} and {upper_bound:.2f}.")
```

```python
# 21 =  Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean ?
# Import required libraries
import numpy as np
from scipy.stats import t

# Step 1: Generate random data from a normal distribution
np.random.seed(123)
true_mean = 70
true_std = 10
sample_size = 25

# Simulate sample data
data = np.random.normal(loc=true_mean, scale=true_std, size=sample_size)

# Step 2: Calculate sample mean and sample standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # ddof=1 for sample std deviation
df = sample_size - 1  # degrees of freedom

# Step 3: Set confidence level and find t-critical value
confidence_level = 0.95
alpha = 1 - confidence_level
t_critical = t.ppf(1 - alpha/2, df)

# Step 4: Calculate margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

# Step 5: Calculate confidence interval
lower = sample_mean - margin_of_error
upper = sample_mean + margin_of_error

# Display results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Sample Std Dev = {sample_std:.2f}")
print(f"Degrees of Freedom = {df}")
print(f"t-critical (two-tailed, 95%) = {t_critical:.3f}")
print(f"Margin of Error = {margin_of_error:.2f}")
print(f"95% Confidence Interval for the Mean: ({lower:.2f}, {upper:.2f})")

# Interpretation
print("\nInterpretation:")
print(f"We are 95% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")
```

```python
# 23 =  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution ?
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Step 1: Define the Poisson distribution parameter (λ = average rate)
lambda_poisson = 4

# Step 2: Define the range of x values (discrete outcomes)
x = np.arange(0, 15)

# Step 3: Calculate CDF values
cdf_values = poisson.cdf(x, mu=lambda_poisson)

# Step 4: Print CDF values
print("x\tP(X ≤ x) (CDF)")
for xi, ci in zip(x, cdf_values):
    print(f"{xi}\t{ci:.4f}")

# Step 5: Plot the CDF
plt.figure(figsize=(8, 4))
plt.step(x, cdf_values, where='post', label='Poisson CDF', color='blue', linewidth=2)
plt.title("Cumulative Distribution Function of Poisson(λ=4)")
plt.xlabel("x")
plt.ylabel("P(X ≤ x)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(x)
plt.ylim(0, 1.1)
plt.legend()
plt.show()
```

```python
# 27 Implement Z-score normalization on a dataset and visualize its transformation ?

# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a sample dataset
np.random.seed(1)
data = np.random.normal(loc=100, scale=20, size=1000)  # mean=100, std=20

# Step 2: Perform Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_normalized = (data - mean) / std

# Step 3: Print before and after stats
print(f"Original Mean: {mean:.2f}, Standard Deviation: {std:.2f}")
print(f"Normalized Mean: {np.mean(z_normalized):.2f}, Standard Deviation: {np.std(z_normalized):.2f}")

# Step 4: Plot original and normalized data
plt.figure(figsize=(12, 5))

# Original data histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='orange', edgecolor='black', alpha=0.7)
plt.title("Original Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True, linestyle="--", alpha=0.6)

# Normalized data histogram
plt.subplot(1, 2, 2)
plt.hist(z_normalized, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Z-score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
# 28 = # Import required libraries
import numpy as np
from scipy.stats import skew, kurtosis

# Function to generate data and compute skewness and kurtosis
def analyze_distribution_stats(size=1000, mean=0, std=1):
    """
    Generates data from a normal distribution and calculates its skewness and kurtosis.
    """
    # Generate normally distributed data
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)  # excess kurtosis (normal dist = 0)

    # Print results
    print(f"Generated Data Size: {size}")
    print(f"Skewness: {data_skewness:.4f}")
    print(f"Kurtosis (excess): {data_kurtosis:.4f}")

    # Interpretation
    print("\nInterpretation:")
    if abs(data_skewness) < 0.5:
        print(" - Skewness is near zero ⇒ distribution is approximately symmetric.")
    else:
        print(" - Distribution is skewed.")

    if abs(data_kurtosis) < 0.5:
        print(" - Kurtosis is near zero ⇒ distribution is mesokurtic (normal-like).")
    elif data_kurtosis > 0:
        print(" - Positive kurtosis ⇒ leptokurtic (heavy tails).")
    else:
        print(" - Negative kurtosis ⇒ platykurtic (light tails).")

# Call the function
analyze_distribution_stats()
```
