#Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

Answer:
  Anomaly Detection is the process of identifying data points, events, or patterns in a dataset that deviate significantly from the majority of the data. These unusual patterns are called anomalies or outliers.
It is widely used in areas such as fraud detection, network security, medical diagnosis, predictive maintenance, and business analytics.

Types of Anomalies

#Point Anomalies

Definition: A data point is considered anomalous if it is far away from the rest of the data.

Example:
In banking transactions, if most daily withdrawals are between ₹1,000 – ₹10,000, but one transaction is ₹5,00,000, that transaction is a point anomaly.

In temperature readings, if the average is 25°C but one reading shows 80°C, that’s a point anomaly.

#Contextual Anomalies (Conditional Anomalies)

Definition: A data point is anomalous only in a specific context (time, location, or situation).

Example:

A credit card bill of ₹20,000 may be normal during the holiday season (Diwali/New Year), but the same amount during a regular month could be an anomaly.

Temperature of 10°C is normal in winter but would be an anomaly in summer.

#Collective Anomalies

Definition: A collection of related data points is anomalous, even if individual points look normal.

Example:

In a network security system, a sudden sequence of login attempts from different countries may indicate a cyber-attack.

In stock markets, if several stocks drop together unusually, it can signal market manipulation or economic shocks.

| Type of Anomaly        | Definition                                             | Example                                                           |
| ---------------------- | ------------------------------------------------------ | ----------------------------------------------------------------- |
| **Point Anomaly**      | Single data point deviates from the rest               | ₹5,00,000 withdrawal in a dataset of ₹1k–10k transactions         |
| **Contextual Anomaly** | Anomaly depends on context (time, location, situation) | 10°C in summer (unusual) but normal in winter                     |
| **Collective Anomaly** | A group of related points is abnormal                  | Multiple failed logins from different IPs (possible cyber-attack) |


#Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

Answer:
#Isolation Forest (iForest)

Approach:

Based on the principle that anomalies are “few and different.”

Randomly selects a feature and splits data. Outliers are easier to isolate → they require fewer splits.

Strengths:

Fast, scalable to large, high-dimensional datasets.

Works well with mixed data (categorical + numerical).

Limitations:

May not perform well if anomalies are not distinctly isolated.

Suitable Use Cases:

Fraud detection in large financial datasets.

Intrusion detection in cybersecurity.

Industrial sensor data (predictive maintenance).

#DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

Approach:

Clustering algorithm based on data density.

Points in low-density areas (not belonging to any cluster) are treated as outliers.

Strengths:

Can find clusters of arbitrary shapes.

Does not require specifying the number of clusters.

Limitations:

Sensitive to parameter choice (eps = neighborhood size, minPts = minimum points per cluster).

Struggles with high-dimensional data.

Suitable Use Cases:

Geographic/spatial anomaly detection (earthquakes, crime hotspots).

Image segmentation.

Detecting noise/outliers in location-based services.

#Local Outlier Factor (LOF)

Approach:

Measures local density deviation of a point relative to its neighbors.

A point is an outlier if it has a much lower density compared to neighbors.

Strengths:

Good for detecting local anomalies in datasets where density varies.

Does not assume global data distribution.

Limitations:

Computationally expensive for large datasets.

Choice of neighbors (k) impacts results.

Suitable Use Cases:

Fraud detection in datasets with regional variations.

Network traffic analysis (detecting unusual patterns in local traffic).

Healthcare anomaly detection (patient vitals with different baselines).

| Algorithm            | Approach                                   | Strengths                                                  | Limitations                                      | Best Use Cases                                          |
| -------------------- | ------------------------------------------ | ---------------------------------------------------------- | ------------------------------------------------ | ------------------------------------------------------- |
| **Isolation Forest** | Random splits → anomalies isolated quickly | Scalable, fast, works in high dimensions                   | Struggles if anomalies are not isolated          | Fraud detection, cybersecurity, predictive maintenance  |
| **DBSCAN**           | Density-based clustering                   | Finds arbitrary-shaped clusters, no need for cluster count | Sensitive to parameters, poor in high-dimensions | Spatial/geographic anomalies, image segmentation        |
| **LOF**              | Local density deviation                    | Good for local anomalies, non-parametric                   | Computationally heavy, sensitive to `k`          | Fraud in regional datasets, network traffic, healthcare |


#Question 3: What are the key components of a Time Series? Explain each with one example.

Answer:
A time series is a sequence of data points recorded at successive, evenly spaced points in time (daily, monthly, yearly, etc.).
To analyze and forecast time series data, we break it into four key components:

#Trend (T)

Definition: The long-term movement or direction in the data over time (upward, downward, or constant).

Example:

Stock prices generally rising over 10 years → upward trend.

Newspaper circulation decreasing steadily over decades → downward trend.

#Seasonality (S)

Definition: Repeating short-term patterns in the data at regular intervals (e.g., daily, monthly, yearly).

Example:

Ice cream sales peak every summer and dip in winter.

Electricity consumption rises every evening (daily seasonality).

#Cyclical Component (C)

Definition: Fluctuations around the trend line that occur over longer, irregular periods, usually linked to business/economic cycles.

Example:

Economic boom and recession cycles every few years.

Real estate prices rising during a housing boom and falling during a crash.

#Irregular/Random Component (I)

Definition: Unpredictable variations caused by random events that cannot be explained by trend, seasonality, or cycles.

Example:

Sudden drop in tourism due to a natural disaster or pandemic.

Unexpected stock market crash due to a political crisis.

| Component            | Meaning                                | Example                           |
| -------------------- | -------------------------------------- | --------------------------------- |
| **Trend**            | Long-term upward or downward direction | Stock prices rising over 10 years |
| **Seasonality**      | Regular repeating patterns             | Ice cream sales peak in summer    |
| **Cyclical**         | Irregular long-term fluctuations       | Economic boom & recession cycles  |
| **Irregular/Random** | Unpredictable variations               | Tourism drop during pandemic      |


#Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?

Answer:  Definition of Stationary Series

A time series is stationary if its statistical properties do not change over time.
That means:

Mean is constant.

Variance is constant.

Covariance between two time periods depends only on the lag, not on actual time.

#Why Stationarity Matters?

Most time series forecasting models (like ARIMA) assume stationarity because it makes the data predictable and mathematically tractable.

#How to Test Stationarity

Visual Inspection (Graphical Test)

Plot the time series.

If you see trend or seasonality, it’s likely non-stationary.

Summary Statistics

Split the data into two parts.

Compare mean and variance in both. If they differ → non-stationary.

Statistical Tests

Augmented Dickey-Fuller (ADF) Test

Null hypothesis (H0): Series is non-stationary.

If p-value < 0.05 → Reject H0 → Series is stationary.

KPSS Test (Kwiatkowski–Phillips–Schmidt–Shin)

Null hypothesis (H0): Series is stationary.

If p-value < 0.05 → Reject H0 → Series is non-stationary.

#How to Transform a Non-Stationary Series into Stationary

Differencing

Subtract current value from previous value.

Removes trend.

Example: y_t - y_(t-1)

Log Transformation

Stabilizes variance (useful when fluctuations increase with level).

Example: Sales data with exponential growth → apply log.

Detrending

Fit a regression line (linear or polynomial) and subtract the trend.

Seasonal Differencing

Subtract the value from the same season in the previous cycle.

Example: y_t - y_(t-12) for monthly data with yearly seasonality.

Moving Average Smoothing

Removes short-term fluctuations and highlights long-term structure.

| Method                    | Purpose                 | Example                      |
| ------------------------- | ----------------------- | ---------------------------- |
| **Differencing**          | Removes trend           | Stock prices trending upward |
| **Log Transformation**    | Stabilizes variance     | Exponential sales growth     |
| **Detrending**            | Removes long-term trend | GDP increasing steadily      |
| **Seasonal Differencing** | Removes seasonality     | Ice cream sales every summer |
| **Smoothing**             | Reduces noise           | Daily temperature variations |

#Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series,and decompose it into trend, seasonality, and residual components.
(Include your Python code and output in the code box below.)


Answer:

  Expected Output (when you run it):

Original Series Plot:

Passenger numbers steadily increase over time (upward trend).

Clear repeating seasonal spikes (summer/holiday travel).

Decomposition Plots:

Trend: Smooth upward line showing long-term growth.

Seasonality: Regular 12-month repeating pattern (high in summer, low in early year).

Residual: Random fluctuations not explained by trend/seasonality.

#Explanation of Results:

Trend: Passenger traffic increased steadily from 1949 to 1960.

Seasonality: Peaks around mid-year (holiday season) and dips at the beginning of the year.

Residual: Small random variations caused by irregular events.

```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Step 2: Load dataset (AirPassengers CSV from seaborn/mldata or manually)
# If you have CSV: df = pd.read_csv("AirPassengers.csv")
# Kaggle version usually has 'Month' and '#Passengers' columns
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url)

# Step 3: Preprocess
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)
df.rename(columns={'Passengers': 'Passengers'}, inplace=True)

# Step 4: Plot original series
plt.figure(figsize=(10, 4))
plt.plot(df['Passengers'], label='AirPassengers Data', color='blue')
plt.title("AirPassengers Time Series")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()

# Step 5: Decompose time series
decomposition = seasonal_decompose(df['Passengers'], model='multiplicative', period=12)

# Step 6: Plot decomposition (trend, seasonality, residual)
fig = decomposition.plot()
fig.set_size_inches(10, 8)
plt.show()
```

#Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.
(Include your Python code and output in the code box below.)


Answer:

Expected Output

Scatter Plot:

Blue points = normal trips (fare increases with distance).

Red points = anomalies (e.g., very high fare for a short trip, or negative fare values).

Interpretation:

If a 2 km trip has $200 fare → anomaly.

If a 30 km trip has $2 fare → anomaly.

```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Step 2: Load a sample NYC Taxi Fare dataset
# (Here we use a smaller public version for demonstration)
url = "https://www.kaggle.com/datasets/neilclack/nyc-taxi-trip-data-google-public-data?select=taxi_trip_data.csv"
df = pd.read_csv(url, nrows=5000)   # load only 5000 rows for speed

# Step 3: Select relevant numerical features
data = df[['fare_amount', 'passenger_count']].dropna()

# Step 4: Fit Isolation Forest
iso = IsolationForest(contamination=0.05, random_state=42)
data['anomaly'] = iso.fit_predict(data[['fare_amount', 'passenger_count']])

# Step 5: Separate normal vs anomalies
normal = data[data['anomaly'] == 1]
anomalies = data[data['anomaly'] == -1]

# Step 6: Plot anomalies
plt.figure(figsize=(8,6))
plt.scatter(normal['fare_amount'], normal['passenger_count'],
            c='blue', label='Normal', alpha=0.5)
plt.scatter(anomalies['fare_amount'], anomalies['passenger_count'],
            c='red', label='Anomaly', alpha=0.7)
plt.xlabel("Fare Amount ($)")
plt.ylabel("Passenger Count")
plt.title("Isolation Forest Anomaly Detection (NYC Taxi Fare)")
plt.legend()
plt.show()
```

#Question 8: Train a SARIMA model on the monthly airline passengers dataset.Forecast the next 12 months and visualize the results.
(Include your Python code and output in the code box below.)

Answer:

#What you’ll see in the output

Blue line: Historical training data (1949–1959).

Green line: Actual passenger values in 1960 (test data).

Red line: SARIMA forecast for 1960.

Pink shaded area: Confidence interval of forecast.

#Interpretation of Results

The SARIMA model captures trend + yearly seasonality well.

Forecasted values closely follow the actual 1960 passenger numbers.

Confidence intervals widen slightly over time (increasing uncertainty).

```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Step 2: Load AirPassengers dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url)

# Convert Month column to datetime
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)

# Step 3: Train-Test Split
train = df[:-12]   # use all but last 12 months for training
test = df[-12:]    # keep last 12 months for testing

# Step 4: Fit SARIMA model
# Order(p,d,q) and Seasonal_order(P,D,Q,m) chosen after ACF/PACF tuning (here we use common config)
model = SARIMAX(train['Passengers'], order=(1,1,1), seasonal_order=(1,1,1,12))
sarima_fit = model.fit(disp=False)

# Step 5: Forecast next 12 months
forecast = sarima_fit.get_forecast(steps=12)
pred_mean = forecast.predicted_mean
conf_int = forecast.conf_int()

# Step 6: Plot actual vs forecast
plt.figure(figsize=(10,6))
plt.plot(train.index, train['Passengers'], label='Train', color='blue')
plt.plot(test.index, test['Passengers'], label='Test (Actual)', color='green')
plt.plot(pred_mean.index, pred_mean, label='Forecast', color='red')
plt.fill_between(pred_mean.index,
                 conf_int.iloc[:,0],
                 conf_int.iloc[:,1], color='pink', alpha=0.3)

plt.title("SARIMA Forecast on AirPassengers Data")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
```

#Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.
(Include your Python code and output in the code box below.)


Answer:  

#Expected Output

Blue points: clustered normal data around (2,2).

Red X markers: scattered anomalies far from the main cluster.

#Interpretation

LOF identifies points in low-density regions (far away from neighbors) as anomalies.

Useful when anomalies are local (relative to neighbors, not global).

```python
# Step 1: Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Step 2: Generate synthetic dataset
np.random.seed(42)

# Normal data (cluster around [2,2])
X_normal = 0.3 * np.random.randn(200, 2) + [2, 2]

# Anomalies (scattered far away points)
X_outliers = np.random.uniform(low=-4, high=8, size=(20, 2))

# Combine dataset
X = np.r_[X_normal, X_outliers]

# Step 3: Apply LOF
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred = lof.fit_predict(X)

# -1 = anomaly, 1 = normal
normal = X[y_pred == 1]
anomalies = X[y_pred == -1]

# Step 4: Visualization
plt.figure(figsize=(8,6))
plt.scatter(normal[:,0], normal[:,1], c='blue', label="Normal Points", alpha=0.6)
plt.scatter(anomalies[:,0], anomalies[:,1], c='red', label="Anomalies", alpha=0.8, marker='x', s=100)
plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

#Question 10: You are working as a data scientist for a power grid monitoring company.

Your goal is to forecast energy demand and also detect abnormal spikes or drops in real-time consumption data collected every 15 minutes. The dataset includes features like timestamp, region, weather conditions, and energy usage.
Explain your real-time data science workflow:

● How would you detect anomalies in this streaming data (Isolation Forest / LOF /
DBSCAN)?

● Which time series model would you use for short-term forecasting (ARIMA /
SARIMA / SARIMAX)?

● How would you validate and monitor the performance over time?

● How would this solution help business decisions or operations?

(Include your Python code and output in the code box below.)


Answer:

#1. Anomaly Detection (Streaming Data)

Choice: Isolation Forest (fast, scalable, works with high-dimensional data).

Why not LOF/DBSCAN?

LOF → better for small/local datasets, but heavy for streaming.

DBSCAN → good for spatial clustering, but not efficient for continuous streaming.

Approach:

Train Isolation Forest on recent sliding window of energy usage.

Flag sudden spikes/drops as anomalies.

#2. Forecasting Model (Short-term)

Choice: SARIMAX

Reason: Energy demand has trend + seasonality (daily/hourly patterns).

SARIMAX can also incorporate exogenous features like weather (temperature, humidity) and region.

Alternative: For very large-scale data, LSTM/Prophet can also be used.

#3. Validation & Monitoring

Validation:

Use train-test split (e.g., last 2 weeks as test set).

Metrics: MAE, RMSE, MAPE for forecasts.

For anomaly detection: Precision, Recall, F1-score (if labeled anomalies available).

Monitoring in Production:

Continuously retrain on new data (online learning).

Use concept drift detection to catch changing patterns in energy consumption.

Set up alerts when anomaly rate > threshold.

#4. Business Impact

Grid Stability: Detect spikes/drops early → prevent blackouts.

Cost Optimization: Forecast demand → optimize energy purchase and distribution.

Maintenance: Detect abnormal drops → identify failing transformers or faulty meters.

Customer Insights: Anticipate peak loads → dynamic pricing or demand-response programs.

#Python Implementation (Simplified Demo)

#Expected Output

Blue line: historical energy demand.

Orange X’s: detected anomalies (spikes/drops).

Red line: SARIMAX forecast for the next hours.

Pink band: forecast confidence intervals.

#Final Note:
This workflow ensures:

Anomaly detection (short-term risk alerts)

Forecasting (resource planning)

Monitoring (continuous improvement)


```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Step 2: Simulate streaming energy dataset
np.random.seed(42)
time_index = pd.date_range("2023-01-01", periods=200, freq="15T")
energy_usage = 100 + 10*np.sin(np.linspace(0, 20, 200)) + np.random.normal(0, 2, 200)

# Inject anomalies (spikes/drops)
energy_usage[50] += 40   # spike
energy_usage[120] -= 30  # drop
energy_usage[180] += 50  # spike

df = pd.DataFrame({"timestamp": time_index, "energy_usage": energy_usage})
df.set_index("timestamp", inplace=True)

# Step 3: Anomaly detection with Isolation Forest
iso = IsolationForest(contamination=0.05, random_state=42)
df["anomaly"] = iso.fit_predict(df[["energy_usage"]])

# Step 4: Forecast with SARIMAX
train = df.iloc[:150]
test = df.iloc[150:]

model = SARIMAX(train["energy_usage"], order=(1,1,1), seasonal_order=(1,1,1,96))
# 96 steps ≈ 1 day (15-min intervals)
results = model.fit(disp=False)

forecast = results.get_forecast(steps=len(test))
pred_conf = forecast.conf_int()

# Step 5: Visualization
plt.figure(figsize=(12,6))
# Historical
plt.plot(df.index, df["energy_usage"], label="Energy Usage", color="blue")
# Forecast
plt.plot(test.index, forecast.predicted_mean, label="Forecast", color="red")
# Confidence Interval
plt.fill_between(pred_conf.index,
                 pred_conf.iloc[:,0],
                 pred_conf.iloc[:,1], color='pink', alpha=0.3)
# Anomalies
plt.scatter(df.index[df["anomaly"]==-1], df["energy_usage"][df["anomaly"]==-1],
            color="orange", label="Anomaly", marker="x", s=100)
plt.title("Real-Time Energy Demand Forecasting & Anomaly Detection")
plt.xlabel("Timestamp")
plt.ylabel("Energy Usage (MW)")
plt.legend()
plt.show()
```
