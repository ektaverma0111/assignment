#1 = What is a RESTful API ?

Ans = A RESTful API (Representational State Transfer Application Programming Interface) is a type of web service that follows the principles and constraints of REST architecture. REST was introduced by Roy Fielding in his doctoral dissertation in 2000 and has since become a standard way to design networked applications.

#2 = Explain the concept of API specification ?

Ans = An API Specification is a formal and structured description of how an Application Programming Interface (API) should behave and how it can be used. It defines what the API does, how to use it, and what responses to expect. In other words, an API specification acts as a contract between the API provider and the consumer (developer).

Key Elements of an API Specification:

1. Endpoints:
An endpoint is a specific URL path where an API can be accessed. Each endpoint is typically associated with a resource or an action.

Example: /users, /products/{id}

2. HTTP Methods:
The specification defines which HTTP methods are allowed for each endpoint, such as:

GET – to retrieve data

POST – to create new data

PUT – to update existing data

DELETE – to remove data

3. Request Format:
It outlines what data must be included in the request. This includes:

Path parameters (e.g., /users/{id})

Query parameters (e.g., /users?age=25)

Request headers

Request body (usually in JSON or XML)

4. Response Format:
Describes the structure and content of the response the API returns. This includes:

Status codes (e.g., 200 OK, 404 Not Found)

Response body format (usually JSON)

Possible error messages

5. Authentication and Authorization:

Specifies how clients should authenticate (e.g., using API keys, OAuth tokens) and what access control rules apply.

6. Data Types and Models:

Defines the structure of data objects used in the API, including data types, required fields, and allowed values.

7. Examples and Descriptions:
Good API specifications often include examples of valid requests and responses, which help developers understand how to use the API correctly.

# 6 = What are HTTP methods used in RESTful APIs ?

Ans = In RESTful APIs, HTTP methods are used to perform various operations on resources. Each method defines a specific kind of action that should be performed on the given resource. These methods are part of the standard HTTP protocol and are fundamental to REST architecture.

1. GET

The GET method is used to retrieve data from the server without making any modifications to the resource. It is a read-only operation and does not affect the resource’s state. This method is safe and idempotent, meaning multiple identical requests will return the same result without changing the resource.

2. POST

The POST method is used to create a new resource on the server. It sends data to the server, and the server processes this data to create a new entry. POST is not idempotent, meaning multiple identical requests may result in multiple resource creations.

3. PUT

The PUT method is used to update or replace an existing resource entirely. If the resource does not exist, it may create one. PUT is idempotent, meaning that sending the same request multiple times results in the same resource state.

4. PATCH

The PATCH method is used to partially update an existing resource. Unlike PUT, which replaces the whole resource, PATCH modifies only specified fields. It is typically idempotent, as repeated identical requests will result in the same changes.

5. DELETE

The DELETE method is used to remove a resource from the server. It deletes the specified resource and is idempotent, since performing the same delete operation multiple times has the same effect—the resource remains deleted.

6. OPTIONS

The OPTIONS method is used to describe the communication options available for a particular resource or server. It informs the client about the allowed HTTP methods and other request options. It does not change the resource state and is used mainly for preflight checks in cross-origin resource sharing (CORS).



#7 What is the purpose of the @app.route() decorator in Flask ?

Ans = Purpose of the @app.route() Decorator in Flask

In Flask, the @app.route() decorator is used to associate a specific URL path with a function, also known as a view function. It tells the Flask application which function should be executed when a client (such as a web browser or API consumer) accesses a particular URL.


Key Purposes of @app.route() Decorator:
1. Defines a Route

It defines a URL pattern that the application can respond to. Each route corresponds to a web address or endpoint.

2. Maps URL to a Function

It connects a URL path to a specific function, which is called automatically when that URL is requested by the client.

3. Handles HTTP Requests

The view function associated with a route handles incoming HTTP requests (such as GET or POST) and returns a response to the client.

4. Supports Multiple Routes

A single function can be bound to multiple routes if needed, allowing different URLs to trigger the same response logic.

5. Enables Dynamic Routing

The @app.route() decorator can include dynamic variables in the URL pattern, allowing the route to handle multiple inputs (like user IDs or names).

6. Defines Allowed HTTP Methods (Optional)

Although it defaults to handling only GET requests, it can be configured to allow other methods like POST, PUT, DELETE, etc., using the methods parameter.

#8 What is the difference between GET and POST HTTP methods ?

Ans = In RESTful APIs and web development, GET and POST are two commonly used HTTP methods. They serve different purposes and have distinct characteristics in how they handle data and interact with resources.

1. Purpose
GET:
Used to retrieve data from the server. It is a read-only operation and does not change the server’s state.

POST:
Used to send data to the server to create a new resource or submit a form. It may change the server’s state.

2. Data Transmission

GET:
Transmits data through the URL, typically using query parameters.
Example format: example.com/page?name=John

POST:
Sends data in the request body, not in the URL. This allows more secure and larger data transmission.

3. Visibility

GET:
Data is visible in the URL, making it less secure and easily bookmarked or cached.

POST:
Data is not visible in the URL, offering better data privacy and security.

4. Use Case

GET:
Ideal for fetching data, reading information, or loading pages.

POST:
Suitable for submitting forms, uploading files, or sending sensitive data.

5. Idempotency

GET:
It is idempotent, meaning multiple identical requests have the same effect (they retrieve the same data).

POST:
It is not idempotent, meaning repeated requests may create multiple entries or have different effects.

6. Caching

GET:
Responses can be cached by browsers and servers for efficiency.

POST:
Responses are not typically cached, since the method changes data on the server.





#9 How do you handle errors in Flask APIs ?

Ans = Error handling in Flask APIs involves managing situations where requests cannot be processed as expected, either due to client mistakes, server issues, or invalid data. Proper error handling improves API reliability, user experience, and helps in debugging.

Key Concepts in Error Handling for Flask APIs:

1. HTTP Status Codes

APIs use HTTP status codes to indicate the outcome of a request.

Common codes for errors include:

400 Bad Request: The client sent an invalid request.

401 Unauthorized: Authentication is required or failed.

403 Forbidden: Client is authenticated but not authorized.

404 Not Found: The requested resource does not exist.

500 Internal Server Error: The server encountered an unexpected condition.

2. Flask’s Built-in Error Handling

Flask provides default responses for common HTTP errors like 404 and 500.

These defaults can be overridden to provide custom error messages or responses.

3. Using @app.errorhandler() Decorator

Flask allows developers to register custom error handlers for specific HTTP status codes or exceptions using the @app.errorhandler() decorator.

Custom handlers return a structured response (e.g., JSON) and an appropriate HTTP status code.

4. Raising Exceptions

Developers can raise HTTP exceptions manually (e.g., abort(404)) to trigger specific error handlers.

This allows early termination of request processing when an error condition is detected.

5. Returning Consistent Error Responses

API error responses should be consistent and informative.

Usually, error responses include:

An error code or status

A human-readable error message

Optionally, details about the error or guidance for resolution

6. Logging Errors

Errors should be logged for monitoring and troubleshooting.

Logging helps track unexpected issues and improve API reliability.

It provides an ORM interface and manages connections between Flask and the database efficiently.

The database connection is configured in the Flask application’s settings using a database URI (Uniform Resource Identifier).

The URI includes information like the database type, username, password, host, port, and database name.

Example format (conceptual):
dialect+driver://username:password@host:port/database

6. Initializing the Database in Flask

The database extension (e.g., Flask-SQLAlchemy) is initialized with the Flask application instance.

This sets up the connection and prepares the application to execute database operations.

7. Defining Models

Database models are Python classes that represent database tables.

Each model’s attributes correspond to table columns.

Models provide an abstraction for CRUD (Create, Read, Update, Delete) operations on the data.

8. Performing Database Operations

Once connected, the application can perform:

Queries to fetch data

Insertions to add new records

Updates to modify existing records

Deletions to remove records

#19 = What are HTTP status codes, and why are they important in a Flask API?

Ans = HTTP status codes are standardized three-digit codes returned by a web server (including Flask APIs) to indicate the result of an HTTP request made by a client. They inform the client about whether the request was successful, encountered an error, or requires further action.

They are part of the HTTP protocol and communicate the outcome of a request.

Status codes are grouped into five classes based on the first digit:

1xx (Informational): Request received, continuing process.

2xx (Success): Request was successfully received, understood, and accepted.

3xx (Redirection): Further action needed to complete the request.

4xx (Client Error): The request contains bad syntax or cannot be fulfilled.

5xx (Server Error): The server failed to fulfill a valid request.

HTTP Status Codes Important in a Flask API:-

1. Communicate Outcome Clearly

They provide clients with a clear understanding of what happened with their request (success, error type, redirect, etc.).

2. Control Client Behavior

Clients can decide how to respond based on the status code (e.g., retry on server error, prompt user on client error).

3. Improve API Usability

Using appropriate status codes makes APIs predictable and easier to use.

4. Facilitate Error Handling

Clients can programmatically handle errors or specific responses using status codes.

5. Conform to Standards

Proper use of HTTP status codes aligns the API with web standards and REST principles.

6. Support Debugging and Monitoring

Status codes help developers quickly identify and diagnose issues.

#21 How would you secure a Flask API ?

Ans = Securing a Flask API involves implementing measures that protect the API from unauthorized access, data breaches, and malicious attacks while ensuring data privacy and integrity.

Key Strategies to Secure a Flask API:

1. Authentication

Ensure that only authorized users or clients can access the API by implementing authentication mechanisms such as:

API Keys: Simple tokens provided to clients.

HTTP Basic Authentication: Username and password with every request (usually over HTTPS).

OAuth / OAuth2: Industry-standard protocol for delegated access.

JWT (JSON Web Tokens): Stateless tokens carrying user claims, commonly used in REST APIs.

2. Authorization

Control what authenticated users are allowed to do by implementing role-based or permission-based access controls.

3. Use HTTPS

Encrypt data in transit using HTTPS to prevent eavesdropping and man-in-the-middle attacks.

4. Input Validation and Sanitization

Validate all incoming data rigorously to prevent injection attacks (e.g., SQL injection, command injection) and other vulnerabilities.

5. Rate Limiting and Throttling

Prevent abuse or denial-of-service (DoS) attacks by limiting the number of requests a client can make in a given time.

6. CORS (Cross-Origin Resource Sharing) Configuration

Properly configure CORS policies to control which domains can access your API resources.

7. Error Handling

Avoid exposing sensitive information in error messages that could aid attackers.

8. Secure Configuration and Secrets Management

Keep secret keys, passwords, and tokens secure and out of source code (e.g., use environment variables).

9. Logging and Monitoring

Log access and errors to detect suspicious activities and respond promptly.

10. Use Security Headers
Implement HTTP security headers like Content-Security-Policy, X-Content-Type-Options, and Strict-Transport-Security.
