# 1 = What is hypothesis testing in statistics ?

Ans = Hypothesis testing is a formal statistical method used to make decisions or inferences about population parameters based on sample data. It helps determine whether there is enough evidence to support a particular belief (hypothesis) about a population.

Key Concepts:

1. Hypothesis:

A hypothesis is a statement or assumption about a population parameter (like the mean or proportion).

2. Types of Hypotheses:

Null Hypothesis (H₀):
It is the default or original assumption. It usually states that there is no effect, no difference, or no relationship.
Example: H₀: μ = 50 (the population mean is 50)

Alternative Hypothesis (H₁ or Ha):
It is what you want to prove. It states that there is an effect, a difference, or a relationship.
Example: H₁: μ ≠ 50

3. Test Statistic:
A value calculated from sample data that is used to test the hypothesis (e.g., z-score or t-score). It measures how far the sample statistic is from the assumed population parameter under the null hypothesis.

4. Significance Level (α):
The threshold for rejecting the null hypothesis. Common values: 0.05 (5%), 0.01 (1%).
It represents the probability of rejecting H₀ when it is actually true (Type I Error).

5. P-value:
The probability of obtaining a test statistic as extreme as the one observed, assuming the null hypothesis is true.

If p-value ≤ α, reject H₀ (significant result)

If p-value > α, fail to reject H₀ (not significant)

6. Decision Rule:
Based on comparing the p-value with α, we decide to either:

Reject the null hypothesis, or

Fail to reject the null hypothesis

Purpose of Hypothesis Testing:

To make data-driven decisions.

To check if observed results are due to chance.

To validate or reject scientific claims or business assumptions.





# 2 = What is the null hypothesis, and how does it differ from the alternative hypothesis ?

Ans = The null hypothesis, symbolized as H₀, is a formal assumption in statistics that states there is no change, no effect, no difference, or no relationship between variables in a population.

It is the starting point in any hypothesis testing process and is assumed to be true until evidence suggests otherwise.

1. Null Hypothesis (H₀):

Definition:

The null hypothesis is a formal statement that there is no effect, no difference, or no relationship between variables.
It represents the status quo, default belief, or commonly accepted fact.

Purpose:
It acts as the starting point for hypothesis testing. We assume it is true until we have enough evidence to reject it.

Examples:

H₀: μ = 50 → The population mean is 50.

H₀: p = 0.4 → The population proportion is 0.4.

H₀: μ₁ = μ₂ → Two population means are equal.

2. 2. Alternative Hypothesis (H₁ or Ha):

Definition:

The alternative hypothesis is a statement that contradicts the null hypothesis. It proposes a real effect, difference, or relationship exists in the population.

Purpose:
It is what the researcher wants to prove or support through data evidence.

Examples:

H₁: μ ≠ 50 → The population mean is not 50.

H₁: p > 0.4 → The population proportion is greater than 0.4.

H₁: μ₁ ≠ μ₂ → Two population means are different.

# 3 = What is the significance level in hypothesis testing, and why is it important ?

The significance level, denoted by the Greek letter α (alpha), is a threshold used in hypothesis testing to determine how much evidence is required to reject the null hypothesis (H₀).

It represents the probability of making a Type I error, which means rejecting the null hypothesis when it is actually true.


Why is the Significance Level Important?

1. Controls Error Risk (Type I Error):

It sets the maximum acceptable probability of wrongly rejecting a true null hypothesis.

If α = 0.05, there is a 5% chance of a false positive (Type I error).

2. Guides Decision Making:
It gives a clear rule:

If p-value ≤ α, reject H₀ (statistically significant).

If p-value > α, fail to reject H₀ (not significant).

3. Reflects Confidence Level:
The significance level is related to confidence level:

α = 0.05 → 95% confidence

α = 0.01 → 99% confidence

4. Affects Sensitivity of the Test:

A lower α means the test is more conservative (less likely to reject H₀).

A higher α increases the risk of error but may detect subtle effects.



# 4 = What does a P-value represent in hypothesis testing ?

Ans = The p-value, or probability value, is a statistical measure that helps determine the strength of the evidence against the null hypothesis (H₀).

It represents the probability of observing the sample results (or more extreme results), assuming that the null hypothesis is true.

Interpretation of the P-value:

A small p-value means the observed result is unlikely under H₀, suggesting evidence against H₀.

A large p-value means the observed result is likely under H₀, suggesting weak or no evidence against H₀.

P-value vs. Significance Level (α):

The p-value is calculated from the data.

The significance level α is chosen before the test.

Decision rule:

If p-value ≤ α → Reject H₀

If p-value > α → Fail to reject H₀

The p-value is not the probability that H₀ is true.

It is also not the probability of making an error.

Instead, it is the probability of getting the observed data (or more extreme) under the assumption that H₀ is correct.

Smaller p-values mean stronger evidence against H₀.

In hypothesis testing, the p-value is a measure that helps determine the strength of the evidence against the null hypothesis (H₀). It represents the probability of obtaining the observed results, or more extreme ones, if the null hypothesis is true. When interpreting the p-value, we compare it to a pre-determined significance level (α), commonly set at 0.05. If the p-value is less than or equal to α, we consider the result statistically significant and reject the null hypothesis, concluding that there is sufficient evidence in favor of the alternative hypothesis (H₁). However, if the p-value is greater than α, we fail to reject the null hypothesis, indicating that the observed data does not provide strong enough evidence to support a real effect or difference. It's important to understand that a small p-value does not prove that the null hypothesis is false, nor does a large p-value prove that it is true—it simply reflects whether the observed outcome is likely or unlikely under the assumption that H₀ is correct. Therefore, the p-value acts as a decision-making tool in statistical inference, guiding researchers in determining whether the results of their sample are consistent with the assumed model or hypothesis.





# 6 =  What are Type 1 and Type 2 errors in hypothesis testing ?

Ans = In hypothesis testing, Type I and Type II errors are two possible mistakes that can occur when making decisions based on sample data. A Type I error occurs when the null hypothesis (H₀) is true, but we incorrectly reject it. This is known as a false positive, and the probability of making this error is represented by the significance level (α), such as 0.05, meaning there's a 5% chance of rejecting H₀ when it is actually correct. On the other hand, a Type II error occurs when the null hypothesis is false, but we fail to reject it. This is called a false negative, and its probability is denoted by β (beta). In this case, we miss detecting a real effect or difference that actually exists. While Type I error risks are controlled by setting the significance level, minimizing Type II errors usually requires increasing the sample size or the power of the test. In simple terms, a Type I error is like sounding a false alarm, while a Type II error is like ignoring a real threat. Both errors are important in statistical decision-making, and there's often a trade-off between them—reducing one can increase the other. Understanding and managing these errors is essential for drawing reliable conclusions in any statistical analysis.



# 7 = What is the difference between a one-tailed and a two-tailed test in hypothesis testing ?

In hypothesis testing, the choice between a one-tailed test and a two-tailed test depends on the direction of the research question or the nature of the alternative hypothesis. A one-tailed test is used when the alternative hypothesis (H₁) states that the population parameter is either greater than or less than a certain value, but not both. This test focuses on only one direction of difference and is more powerful in detecting effects in that specific direction. For example, if a researcher believes that a new drug will increase recovery rates, the test would only look for values greater than the standard, making it a right-tailed test. Conversely, a left-tailed test is used when expecting a decrease.

In contrast, a two-tailed test is applied when the alternative hypothesis suggests that the population parameter is different from the stated value, without specifying a direction—meaning it could be either higher or lower. This type of test divides the significance level (α) equally between both ends (tails) of the distribution. For example, if testing whether a new teaching method leads to any change in test scores (either improvement or decline), a two-tailed test is appropriate.

The key difference lies in the direction of the test: a one-tailed test is directional, while a two-tailed test is non-directional. The choice affects both the interpretation of results and the statistical power. Using a one-tailed test when a two-tailed is required (or vice versa) can lead to misleading conclusions. Therefore, selecting the appropriate test depends on the specific research hypothesis and whether the expected effect has a predicted direction.

# 8 = What is the Z-test, and when is it used in hypothesis testing ?

Ans = A Z-test is a statistical method used in hypothesis testing to determine whether there is a significant difference between sample data and a population parameter when the population standard deviation (σ) is known. It is based on the standard normal distribution, also called the Z-distribution, which has a mean of 0 and a standard deviation of 1. The Z-test calculates a Z-score, which measures how many standard deviations the sample statistic is away from the hypothesized population parameter under the null hypothesis. This Z-score is then compared to a critical value determined by the chosen significance level (α), or it is used to compute the p-value.

The Z-test is typically used when the sample size is large (n ≥ 30) and the population variance is known or reliably estimated. It is especially useful when testing hypotheses about population means or proportions. For example, if a researcher wants to test whether the average height of adult males in a city differs from the national average, and the population standard deviation is known, a Z-test would be appropriate. Depending on the research question, the test may be one-tailed (if the direction of difference is specified) or two-tailed (if only a difference is being tested without a direction).

In summary, the Z-test is applied in situations where the normality assumption holds, the sample size is sufficiently large, and the population standard deviation is known. It provides a simple yet powerful tool for making decisions about population parameters based on sample evidence.



# 9 = How do you calculate the Z-score, and what does it represent in hypothesis testing ?

Interpretation of Z-score:

A Z-score of 0 indicates the data point is exactly at the mean.

A positive Z-score indicates the data point is above the mean.

A negative Z-score indicates the data point is below the mean.

The magnitude of the Z-score shows how far away the point is from the mean in terms of standard deviation.

Z-score in Hypothesis Testing:

In hypothesis testing, the Z-score helps determine whether to reject the null hypothesis.

Role in Hypothesis Testing:

It is used when the sampling distribution is normal (or approximates normal) and the population standard deviation is known.

The Z-score calculates how far the sample statistic (like the sample mean) is from the null hypothesis value, in units of standard error.

Purpose of Using Z-score:

It helps to find the probability (or p-value) of obtaining a test statistic as extreme as the observed one.

Based on this probability, we can determine whether to reject or fail to reject the null hypothesis.

In hypothesis testing, the Z-score acts as a bridge between the sample data and the theoretical distribution, allowing us to assess whether the observed results are statistically significant or due to random chance.

A 95% confidence interval means that if we were to take many random samples and build a confidence interval from each, about 95% of those intervals would contain the true population parameter.

A confidence interval is a fundamental concept in inferential statistics that allows researchers to estimate population parameters while expressing the degree of uncertainty. # 16 = How is Bayes' Theorem used in statistics, and what is its significance ?

Bayes' Theorem is a fundamental principle in probability and statistics that describes how to update the probability of a hypothesis based on new evidence or information. It provides a mathematical framework for revising prior beliefs in light of observed data.

How It Is Used in Statistics:

1. Updating Beliefs with Evidence:
Bayes’ theorem allows statisticians to revise prior probabilities (initial assumptions) after observing new data.

It turns prior information into a posterior distribution, reflecting both prior knowledge and new evidence.

2. Bayesian Inference:
In Bayesian statistics, Bayes’ theorem is the foundation for all inference.

Unlike classical (frequentist) statistics, Bayesian methods treat parameters as random variables and produce probability distributions for them.

3. Hypothesis Testing:
It helps calculate the probability of a hypothesis given the data (posterior), rather than calculating the probability of observing data under a hypothesis.

4. Classification Problems:
Bayes’ theorem is used in Naive Bayes classifiers, a popular machine learning algorithm for categorization tasks (e.g., spam detection, sentiment analysis).

5. Medical Diagnostics:
It is widely applied to estimate the probability of a disease given test results, accounting for false positives and negatives.

Significance of Bayes’ Theorem:

1. Combines Prior Knowledge and Evidence:

It allows statisticians to incorporate existing beliefs or expert knowledge into the analysis, improving accuracy.

2. Handles Uncertainty Naturally:

Bayesian methods provide full probability distributions, offering more flexible and intuitive interpretations than point estimates alone.

3. Foundation of Bayesian Statistics:

Bayes’ theorem forms the core of Bayesian inference, which is increasingly used in modern statistical analysis and machine learning.

4. Improves Decision Making:

It supports adaptive learning—as new data comes in, predictions and estimates can be updated, leading to more informed decisions.

Bayes’ Theorem is a powerful tool in statistics that allows for the dynamic updating of probabilities as new data becomes available. Its ability to combine prior beliefs with new evidence makes it central to Bayesian reasoning, influencing fields ranging from data science and medicine to economics and artificial intelligence.

# 21 =  What are the different types of ANOVA tests ?

Ans = ANOVA (Analysis of Variance) is a family of statistical tests used to compare means across multiple groups. Depending on the number of factors (independent variables) and whether the same subjects are measured across conditions, there are several types of ANOVA. Each type serves a different experimental design and research question.

1. One-Way ANOVA
Purpose:
To compare the means of three or more independent groups based on one categorical independent variable (factor).

Example:
Testing whether the average test scores differ among students from three different schools.

Design:
One independent variable, with two or more levels (e.g., School A, School B, School C)

One dependent variable (e.g., test score)

2. Two-Way ANOVA (Factorial ANOVA)
Purpose:
To examine the effect of two independent variables (factors) on a single dependent variable, including their interaction effect.

Example:
Analyzing how teaching method (traditional vs. modern) and gender (male vs. female) affect students' performance.

Design:
Two factors, each with two or more levels

Can analyze:

Main effect of Factor A

Main effect of Factor B

Interaction effect (A × B)

3. Repeated Measures ANOVA
Purpose:
Used when the same subjects are measured multiple times under different conditions or at different time points.

Example:
Measuring blood pressure of patients before, during, and after treatment.

Design:
Within-subjects design

One group tested across multiple conditions

Accounts for correlations within subjects

4. Two-Way Repeated Measures ANOVA
Purpose:
Tests the effect of two within-subject factors or one within-subject and one between-subject factor on a dependent variable.

Example:
Studying the effect of time (morning, evening) and treatment type on anxiety levels, measured repeatedly in the same participants.

5. Mixed-Design ANOVA (Split-Plot ANOVA)
Purpose:
Combines between-subjects and within-subjects factors in a single analysis.

Example:
Measuring test scores of two different groups (e.g., control vs. experimental) across three time points (pre-test, mid-test, post-test).

6. MANOVA (Multivariate Analysis of Variance)
Purpose:
An extension of ANOVA used when there are two or more dependent variables to test simultaneously.

Example:
Testing whether teaching style affects both math scores and reading scores together.

7. ANCOVA (Analysis of Covariance)
Purpose:
Combines ANOVA with regression, allowing the adjustment of covariates (continuous variables) that may influence the dependent variable.

Example:
Comparing test scores across schools while controlling for students’ IQ as a covariate.

```python
# 15  Implement hypothesis testing using Z-statistics for a sample dataset ?
import numpy as np
from scipy.stats import norm

# Step 1: Sample data (e.g., test scores)
sample_data = np.array([72, 75, 78, 71, 74, 77, 79, 73, 76, 80])
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Step 2: Known population parameters
population_mean = 70      # μ₀ (null hypothesis mean)
population_std = 8        # σ (population standard deviation)

# Step 3: Calculate Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 4: Calculate p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Step 5: Significance level
alpha = 0.05

# Step 6: Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 7: Hypothesis test conclusion
if p_value < alpha:
    print("Result: Reject the null hypothesis (significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")
```

```python
# 24 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.?
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_skewness_kurtosis(mean=0, std_dev=1, size=1000):
    """
    Generates data from a normal distribution and calculates its skewness and kurtosis.
    """
    # Step 1: Generate normal data
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Step 2: Calculate skewness and kurtosis
    skewness = skew(data)
    kurt = kurtosis(data)  # By default, it returns excess kurtosis

    # Step 3: Print results
    print(f"Mean: {mean}, Std Dev: {std_dev}, Sample Size: {size}")
    print(f"Skewness: {skewness:.4f} (Expected ≈ 0 for normal)")
    print(f"Kurtosis (Excess): {kurt:.4f} (Expected ≈ 0 for normal)")

    # Interpretation
    if abs(skewness) < 0.5:
        print(" → The data is approximately symmetric.")
    elif skewness > 0:
        print(" → The data is positively skewed (right-tailed).")
    else:
        print(" → The data is negatively skewed (left-tailed).")

    if abs(kurt) < 1:
        print(" → The data has normal-like tails (mesokurtic).")
    elif kurt > 0:
        print(" → The data has heavy tails (leptokurtic).")
    else:
        print(" → The data has light tails (platykurtic).")

# Example usage
analyze_skewness_kurtosis()
```

```python
#1 = Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results ?
import numpy as np
from scipy.stats import norm

# Step 1: Define the sample data
sample_data = np.array([72, 75, 78, 71, 74, 77, 79, 73, 76, 80])
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Step 2: Define population parameters
population_mean = 70     # Null hypothesis mean (μ₀)
population_std = 8       # Known population standard deviation (σ)

# Step 3: Calculate the Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 4: Calculate the p-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Step 5: Set the significance level
alpha = 0.05

# Step 6: Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 7: Interpret the result
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis.")
    print("→ The sample mean is significantly different from the population mean.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
    print("→ The sample mean is not significantly different from the population mean.")
```

```python
# 5  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing ?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    """
    Calculates and visualizes Type I and Type II errors for right-tailed Z-test.
    """

    # Standard error
    se = sigma / np.sqrt(n)

    # Critical z and critical x̄
    z_crit = norm.ppf(1 - alpha)
    x_crit = mu0 + z_crit * se

    # Type II error (β)
    z_beta = (x_crit - mu1) / se
    beta = norm.cdf(z_beta)
    power = 1 - beta

    # Generate x values and distributions
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    y0 = norm.pdf(x, mu0, se)  # Under H0
    y1 = norm.pdf(x, mu1, se)  # Under H1

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label="H₀: μ = {}".format(mu0), color="blue")
    plt.plot(x, y1, label="H₁: μ = {}".format(mu1), color="green")

    # Shade Type I error (α) in H₀ distribution
    plt.fill_between(x, y0, where=(x > x_crit), color="red", alpha=0.4, label="Type I Error (α)")

    # Shade Type II error (β) in H₁ distribution
    plt.fill_between(x, y1, where=(x < x_crit), color="orange", alpha=0.4, label="Type II Error (β)")

    # Critical line
    plt.axvline(x_crit, color="black", linestyle="--", label=f"Critical Value = {x_crit:.2f}")

    # Labels and legend
    plt.title("Visualization of Type I and Type II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    # PRINT OUTPUT
    print("🔹 Critical Value (x̄):", round(x_crit, 2))
    print("🔹 Standard Error (SE):", round(se, 2))
    print("🔹 Z-critical (for α):", round(z_crit, 2))
    print("🔹 Z-value for β:", round(z_beta, 2))
    print("🔹 Type I Error (α):", round(alpha, 2))
    print("🔹 Type II Error (β):", round(beta, 4))
    print("🔹 Power of the Test (1 - β):", round(power, 4))

```

```python
# 9  Write a Python function to calculate the confidence interval for a sample mean and explain its significance ?
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for the sample mean using the t-distribution.

    Parameters:
    - data (array-like): Sample data
    - confidence (float): Confidence level (default = 0.95 for 95%)

    Returns:
    - (lower_bound, upper_bound): Tuple representing the confidence interval
    """

    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_critical * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Sample Mean: {mean:.2f}")
    print(f"Sample Size: {n}")
    print(f"Standard Error: {std_err:.4f}")
    print(f"T-critical (df={n-1}): {t_critical:.4f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    return (lower_bound, upper_bound)

    # Simulate some sample data
np.random.seed(0)
sample_data = np.random.normal(loc=100, scale=15, size=30)

# Call the function
confidence_interval(sample_data, confidence=0.95)
```

```python
# 10  Write a Python program to calculate the margin of error for a given confidence level using sample data ?
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence_level=0.95):
    """
    Calculates the margin of error (MoE) for a sample mean using the t-distribution.

    Parameters:
    - data: List or array of sample values
    - confidence_level: Confidence level for the interval (default = 0.95)

    Returns:
    - Margin of error (float)
    """
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # Sample standard deviation
    se = std_dev / np.sqrt(n)       # Standard error

    t_critical = t.ppf((1 + confidence_level) / 2, df=n-1)
    margin_of_error = t_critical * se

    # Output
    print(f"Sample Mean: {mean:.2f}")
    print(f"Sample Size: {n}")
    print(f"Standard Deviation: {std_dev:.2f}")
    print(f"Standard Error: {se:.4f}")
    print(f"T-critical (df={n-1}): {t_critical:.4f}")
    print(f"Confidence Level: {confidence_level * 100:.0f}%")
    print(f"Margin of Error: ±{margin_of_error:.2f}")

    return margin_of_error

    # Sample data
sample_data = [72, 75, 78, 74, 70, 76, 77, 79, 73, 80]

# Call the function
calculate_margin_of_error(sample_data, confidence_level=0.95)
```

```python
# 11  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process ?
def bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A):
    """
    Applies Bayes' Theorem:
    P(A | B) = (P(B | A) * P(A)) / [(P(B | A) * P(A)) + (P(B | ¬A) * P(¬A))]
    """
    numerator = likelihood_B_given_A * prior_A
    denominator = (likelihood_B_given_A * prior_A) + (likelihood_B_given_not_A * prior_not_A)
    posterior = numerator / denominator
    return posterior

# Define the inputs
P_D = 0.01                     # Prior: P(Disease)
P_Pos_given_D = 0.95           # Likelihood: P(Positive | Disease)
P_not_D = 1 - P_D              # Complement: P(No Disease)
P_Pos_given_not_D = 0.05       # Likelihood: P(Positive | No Disease)

# Calculate the posterior probability
P_D_given_Pos = bayes_theorem(P_D, P_Pos_given_D, P_not_D, P_Pos_given_not_D)

# Print the result
print(f"Probability of having disease given positive test (posterior): {P_D_given_Pos:.4f} or {P_D_given_Pos*100:.2f}%")
```

```python
# 12  Perform a Chi-square test for independence between two categorical variables in Python ?
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create a contingency table
# Example: Survey data of gender vs product preference
data = {
    'Product A': [30, 20],  # [Males, Females]
    'Product B': [10, 40]
}
# Rows = Gender (Male, Female), Columns = Products
contingency_table = pd.DataFrame(data, index=['Male', 'Female'])

print("Contingency Table:")
print(contingency_table)

# Step 2: Perform the Chi-square test
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

# Step 3: Print the results
print("\nChi-square Statistic:", round(chi2_stat, 4))
print("Degrees of Freedom:", dof)
print("P-value:", round(p_val, 4))
print("\nExpected Frequencies (if independent):")
print(pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns))

# Step 4: Interpret the result
alpha = 0.05
if p_val < alpha:
    print("\nConclusion: Reject the null hypothesis.")
    print("→ There is a significant association between gender and product preference.")
else:
    print("\nConclusion: Fail to reject the null hypothesis.")
    print("→ Gender and product preference are likely independent.")
```

```python
# 14 Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution ?

import numpy as np
from scipy.stats import chisquare

# Step 1: Observed frequencies from experiment
observed = np.array([8, 9, 10, 11, 12, 10])

# Step 2: Expected frequencies (uniform for a fair 6-sided die)
expected = np.array([10, 10, 10, 10, 10, 10])  # Total rolls = 60, expected per side = 10

# Step 3: Perform Chi-square Goodness-of-Fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Step 4: Display results
print("Chi-square Statistic:", round(chi2_stat, 4))
print("P-value:", round(p_value, 4))

# Step 5: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis.")
    print("→ The observed frequencies do NOT match the expected distribution.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
    print("→ The observed frequencies match the expected distribution.")
```
